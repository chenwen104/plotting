```python
from ggplot import *
```

```python
%matplotlib notebook
```

```python
ggplot( aesthetics= aes(x = 'weight', y = 'hindfoot_length'), data = surveys_complete)
```

```python
ggplot( aesthetics= aes(x = 'weight', y = 'hindfoot_length'), data = surveys_complete) + geom_point()

```

```python
ggplot( aes(x = 'weight', y = 'hindfoot_length'), data = surveys_complete) + geom_point(alpha = 0.1, color = 'blue')
#alpha add indensity if data are on top of each other

```

```python
ggplot( aes(x = 'species_id', y = 'hindfoot_length'), data = surveys_complete) + geom_boxplot()
```

```python
ggplot(aes(x = 'weight', y = 'hindfoot_length', color='species_id'),data = surveys_complete) + \
    geom_point( alpha = 0.1)
```
